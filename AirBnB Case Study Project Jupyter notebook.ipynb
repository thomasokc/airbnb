```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
n_groups = air_bnb['neighbourhood_group']
print(n_groups.value_counts())

print(f"There are {n_groups.value_counts().count()} groups, {n_groups.value_counts().keys()[0]} is the most frequent")
```

```python
# Are private rooms the most popular in manhattan?
r_type = air_bnb['room_type']
display(r_type.value_counts())

m_pop = air_bnb[air_bnb['neighbourhood_group']=='Manhattan']
m_pop

test = m_pop['room_type']
display(test.value_counts())

print(f"The most popular room type is: {test.value_counts().keys()[0]}, so false")
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb.groupby('host_name').sum().sort_values('number_of_reviews', ascending = False).head()
```

```python
#Which neighorhood group has the highest average price?
print('Manhattan')

air_bnb.groupby('neighbourhood_group').mean().sort_values('price', ascending = False).head()
```

```python
# Which neighbor hood group has the highest total price?

air_bnb.groupby('neighbourhood_group').sum().sort_values('price', ascending = False).head()
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby('host_name').sum().sort_values('price', ascending = False).head()
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb[air_bnb.number_of_reviews > 100].groupby('host_name').sum().sort_values('availability_365',ascending = True).head(20)
```

```python
# What host has the highest total of prices and where are they located?
air_bnb.groupby('host_name').sum().sort_values('price', ascending = False).head()
```

```python
# When did Danielle from Queens last receive a review?
air_bnb[air_bnb.host_name == 'Danielle'][air_bnb.neighbourhood_group == 'Queens'].sort_values('last_review', ascending = True)
```
